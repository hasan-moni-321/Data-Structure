```python
class Node(object): 
    def __init__(self, data): 
        self.left = None
        self.data = data 
        self.right = None
        self.height = 1
  

class AVL_Tree(object): 

    def inserting_data(self, root, value): 
      
        # Binary Search Tree
        if not root: 
            return Node(value) 
        elif value < root.data: 
            root.left = self.inserting_data(root.left, value) 
        else: 
            root.right = self.inserting_data(root.right, value) 
  

        root.height = 1 + max(self.finding_height(root.left), self.finding_height(root.right)) 
  
        balance = self.finding_difference(root) 
  
        #Case 1 - Left Left
        if balance > 1 and value < root.left.data: 
            return self.right_rotate(root) 
  
        # Case 2 - Right Right 
        if balance < -1 and value > root.right.data: 
            return self.left_rotate(root) 
  
        # Case 3 - Left Right 
        if balance > 1 and value > root.left.data: 
            root.left = self.left_rotate(root.left) 
            return self.right_rotate(root) 
  
        # Case 4 - Right Left 
        if balance < -1 and value < root.right.data: 
            root.right = self.right_rotate(root.right) 
            return self.left_rotate(root) 
  
        return root 
  
    def left_rotate(self, z): 
  
        y = z.right 
        T2 = y.left 
  
        # Perform rotation 
        y.left = z 
        z.right = T2 
  
        # Update heights 
        z.height = 1 + max(self.finding_height(z.left), self.finding_height(z.right)) 
        y.height = 1 + max(self.finding_height(y.left), self.finding_height(y.right)) 
  
        # Return the new root 
        return y 
  
    def right_rotate(self, z): 
  
        y = z.left 
        T3 = y.right 
  
        # Perform rotation 
        y.right = z 
        z.left = T3 
  
        # Update heights 
        z.height = 1 + max(self.finding_height(z.left), self.finding_height(z.right)) 
        y.height = 1 + max(self.finding_height(y.left), self.finding_height(y.right)) 
  
        # Return the new root 
        return y 
  
    def finding_height(self, root): 
        if not root: 
            return 0
  
        return root.height 
  
    def finding_difference(self, root): 
        if not root: 
            return 0
  
        return self.finding_height(root.left) - self.finding_height(root.right) 
  
    def inorder_print(self, root): 
  
        if not root: 
            return
   
        self.inorder_print(root.left) 
        print(root.data, end=' ')
        self.inorder_print(root.right) 
  
  
# Driver program to test above function 
tree = AVL_Tree() 
root = None

#inserting data
root = tree.inserting_data(root, 80) 
root = tree.inserting_data(root, 150) 
root = tree.inserting_data(root, 5) 
root = tree.inserting_data(root, 87) 
root = tree.inserting_data(root, 52) 
root = tree.inserting_data(root, 4) 
root = tree.inserting_data(root, 0) 
root = tree.inserting_data(root, 17) 
root = tree.inserting_data(root, 32) 
root = tree.inserting_data(root, 40) 
  

  
#printing the data
tree.inorder_print(root)
  
```
