```python
class Node():
    def __init__(self, data):
        self.data = data
        self.next=None
        
class Dequeue():
    def __init__(self):
        self.head=None
        
        
    def print_deque(self):
        value = self.head
        
        while value is not None:
            print(value.data)
            value=value.next
    
    
    def adding_at_beginning(self, newnode):    
        newnode.next = self.head
        self.head = newnode
        
        
    def adding_at_end(self, newnode):
        
        if self.head is None:
            self.head = newnode 
            return
            
        last = self.head      
        while (last.next):
            last = last.next
        last.next = newnode
        
        
    def deleting_from_beginning(self):
        headval = self.head
        
        if headval is not None:
            self.head = headval.next
            headval=None
            
    def deleting_from_end(self):
        last = self.head
        
        while (last.next.next):
            last = last.next
        last.next=None
        
        
de = Dequeue()
de.head=Node('One')

e2 = Node('Two')
e3 = Node('Three')
e4 = Node('Four')
e5 = Node('Five')

de.head.next = e2
e2.next = e3
e3.next = e4
e4.next = e5

#printing value of the deque
print('Initia element of the deque : ')
de.print_deque()

#adding at the beginning
de.adding_at_beginning(Node('zero'))

#printing after adding zero
print('Adding value zero at beginning : ')
de.print_deque()

#adding at the end
de.adding_at_end(Node('six'))

#printing after adding six
print('Adding value siz at end : ')
de.print_deque()

#delete a data from the beginning 
de.deleting_from_beginning()

print('After deleting from beginning : ')
de.print_deque()

#delete a data from the end 
de.deleting_from_end()

print('After deleting from end :')
de.print_deque()
```

```python
import collections  as c

deque = c.deque([1,3,5,7,9,7])

print('The initial deque is :',deque)

#adding 11 at the end
deque.append(11)
print('After adding 11 at the end :',deque)

#adding 0 at the beginning
deque.appendleft(0)
print('After adding 0 at the beginning :',deque)

#deleting 11 from the last
deque.pop()
print('After deleting last element :',deque)

#deleting 1 form the beginning
deque.popleft()
print('After deleting first element :',deque)

#finding the index of the 7 value
print('The index of 7 is :', deque.index(7))

#counting total number of 5
print('The total number of 5 is :',deque.count(5))

#removing a specific number
deque.remove(3)
print('After removing the 3 :',deque)

#extending deaue in right
deque.extend([8,9,10,11])
print('After extending the deque in right:',deque)

#extending deque in left
deque.extendleft([2,4,6])
print('After extending the deque in left:', deque)

#rotating deque
deque.rotate(-3)
print('After rotating deque :',deque)

deque.reverse()
print('After reversing of the deque :',deque)
```
