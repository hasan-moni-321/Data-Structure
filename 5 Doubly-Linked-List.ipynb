```python
import gc

class make_node():
    def __init__(self, data=None):
        self.data=data
        self.prev=None
        self.next=None

class Doubly_Linked_List():
    def __init__(self):
        self.head=None
        
    def printing_Doubly_Linked_List(self):
        value = self.head
        while (value is not None):
            print(value.data)
            value=value.next
            
    def push_at_the_beginning(self, data):
        new_node = make_node(data)
        
        new_node.next=self.head
        
        if self.head is not None:
            self.head.prev=new_node         
        self.head=new_node
        
    
    def adding_at_the_middle(self, prev_node, data):
        if prev_node is None:
            print('The given node is not here')
            return
        
        new_node = make_node(data)
        
        new_node.next = prev_node.next
        
        prev_node.next = new_node
        new_node.prev = prev_node
        
        if new_node.next is not None:
            new_node.next.prev = new_node
            
    
    def appending_at_the_end(self, data):
        new_node = make_node(data)
        
        new_node.next=None
        
        if self.head is None:
            self.head.prev=None
            self.head=new_node
            return
        
        last = self.head
        while last.next is not None:
            last = last.next
        
        last.next=new_node
        new_node.prev=last
        
    def dropping_a_node(self, del_data):
        
        temp = self.head
        
        if temp is not None:
            if (temp.data==del_data):
                self.head = temp.next
                temp=None
                return
            
        while(temp is not None):
            if temp.data==del_data:
                break
            prev=temp
            temp=temp.next
            
        if temp==None:
            return
        
        prev.next = temp.next
        temp=None
   

    
dll = Doubly_Linked_List()
dll.head = make_node('one')

e2 = make_node('two')
e3 = make_node('three')
e4 = make_node('four')
e5 = make_node('five')

dll.head.next=e2
e2.next = e3
e3.next = e4
e4.next = e5

#printing Doubly Linked List
print('Initial value of the Linked List :')
dll.printing_Doubly_Linked_List()

#adding value at the beginning of the list
dll.push_at_the_beginning('zero')
print('\n\nPrinting value after adding zero :')
dll.printing_Doubly_Linked_List()

#adding value at the middle of the list
dll.adding_at_the_middle(dll.head.next.next.next,'three and half')
print('\n\nPrinting value after adding three and half :')
dll.printing_Doubly_Linked_List()

#adding value at the end of the list
dll.appending_at_the_end('six')
print('\n\nPrinting value after adding six :')
dll.printing_Doubly_Linked_List()

#delete node
dll.dropping_a_node('two')
print('\n\nPrinting value after dropping two :')
dll.printing_Doubly_Linked_List()
```
