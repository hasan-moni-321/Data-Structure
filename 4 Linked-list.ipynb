```python
class making_node():
    def __init__(self, data=None):
        self.data = data
        self.next = None
        
class LinkedList():
    def __init__(self):
        self.head = None
        
    def printing_linked_list(self):
        val = self.head
        while val is not None:
            print(val.data)
            val = val.next
            
    def adding_at_beginning(self, val):
        newnode = making_node(val)
        
        newnode.next=self.head
        self.head = newnode
        
    def adding_in_middle(self, oldval, newval):
        if oldval is None:
            print('doesnt have the value',oldval)
            return
        
        val = making_node(newval)
        val.next = oldval.next
        oldval.next = val
        
    def adding_at_the_end(self, newnode):
        newnode = making_node(newnode)
        if self.head is None:
            self.head = newnode
            return
        last = self.head
        while(last.next):
            last = last.next
        last.next = newnode
        
    def removing_a_node(self, value):
        temp = self.head
        
        if temp is not None:
            if (temp.data==value):
                self.head = temp.next
                temp=None
                return
            
        while(temp is not None):
            if temp.data==value:
                break
            prev=temp
            temp=temp.next
            
        if temp==None:
            return
        
        prev.next = temp.next
        temp=None
        
        
    
ll = LinkedList()
ll.head = making_node('Bangladesh')

#making node
ele2 = making_node('India')
ele3 = making_node('Pakistan')
ele4 = making_node('Srilanka')
ele5 = making_node('Vutan')

#linked with one other node
ll.head.next = ele2
ele2.next = ele3
ele3.next = ele4
ele4.next = ele5

#printing value
print('Values of the linked list : ')
ll.printing_linked_list()

#adding vlaue at beginning
ll.adding_at_beginning('maldip')

#print value after adding value at the beginning
print('\n\n','Values after adding at the beginning : ')
ll.printing_linked_list()

#adding value in the middle
ll.adding_in_middle(ll.head.next.next, 'America')

#print value after adding value at the middle
print('\n\n','Values after adding at the middle : ')
ll.printing_linked_list()

#adding Germany at last
ll.adding_at_the_end('Germany')

#print value after adding Germany at last
print('\n\n','Values after adding at the end : ')
ll.printing_linked_list()

#removing a node
ll.removing_a_node('Pakistan')

#print value after removing Pakistan
print('\n\n','Values after removing value Pakistan : ')
ll.printing_linked_list()
```
